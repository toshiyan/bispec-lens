Plot cross-skew spectra

```python
# external modules
import numpy as np
from matplotlib.pyplot import *
# moduels from cmblensplus/utils/
import plottools as pl
```

Initialize parameters

```python
D = 'data/skewspec/'
zss = [0.5078,1.0334,1.5345,2.0548,1100]
c = ['r','g','b','m','c']
```

Plot both theoretical and simulated skew-spectra

```python
for i, zs0 in enumerate(zss):
    pl.plot_1dstyle(ylab=r'$L\times S_L$',xmin=100,xmax=2000,ymin=8e-10,ymax=7e-8,ylog=True)
    for j, zs1 in enumerate(zss):
        if i == j: continue
        L, sbl = np.loadtxt(D+'sim_nres13_zs'+str(zs0)+'x'+str(zs1)+'_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
        b, pbl = np.loadtxt(D+'Sl_RT_zs'+str(zs0)+'x'+str(zs1)+'_zn100_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
        b, PBl = np.loadtxt(D+'Sl_RT_zs'+str(zs0)+'x'+str(zs1)+'_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
        plot(L,L*sbl,color=c[j],alpha=0.3)
        plot(b,b*PBl,color=c[j])
        plot(b,b*pbl,label=str(zs0)+' x '+str(zs1),lw=1,ls='--',color=c[j])
    legend(loc=0,ncol=2)
    show()
    clf()
```
