```python
import numpy as np, plottools as pl
from matplotlib.pyplot import *
D = 'data/skewspec/'
zss = [0.5078,1.0334,1.5345,2.0548,1100]
c = ['r','g','b','m']
```

Galaxy shear for each zs w and w/o post-Born

```python
pl.plot_1dstyle(ylab=r'$L\times S_L$',xmin=100,xmax=2000,ymin=8e-10,ymax=2e-8,ylog=True)
for i, zs in enumerate(zss[:4]):
    L, sbl = np.loadtxt(D+'sim_nres13_zs'+str(zs)+'_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
    l, tbl = np.loadtxt(D+'Sl_RT_zs'+str(zs)+'_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
    b, pbl = np.loadtxt(D+'Sl_RT_zs'+str(zs)+'_zn100_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
    plot(L,L*sbl,color=c[i],alpha=0.3)
    plot(l,l*tbl,label='$z_s='+str(zs)+'$',lw=1,color=c[i])
    plot(b,b*pbl,label='',lw=1,ls='--',color=c[i])
text(200,1e-9,'Dashed: w/o post-Born')
legend(loc=0,ncol=2)
#savefig('fig_skewspec_zs.png')
```

Comparison between fitting formulas

```python
pl.plot_1dstyle(ylab=r'$L\times S_L$',xmin=100,xmax=2000,ymin=1e-9,ymax=2e-8,ylog=True)
for i, zs in enumerate(zss[:4]):
    L, sbl = np.loadtxt(D+'sim_nres13_zs'+str(zs)+'_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
    l, tbl = np.loadtxt(D+'Sl_RT_zs'+str(zs)+'_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
    b, pbl = np.loadtxt(D+'Sl_GM_zs'+str(zs)+'_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
    plot(L,L*sbl,color=c[i],alpha=0.3)
    plot(l,l*tbl,label='$z_s='+str(zs)+'$',lw=1,color=c[i])
    plot(b,b*pbl,label='',lw=1,ls='--',color=c[i])
text(200,1.2e-9,'Dashed: 9-params (Gil-Marin et al. 2012)')
legend(loc=0,ncol=2)
```

Changing Minimum multipole of the convergence map

```python
pl.plot_1dstyle(ylab=r'$L\times S_L$',xmin=200,xmax=2000,ymin=8e-10,ymax=2e-8,ylog=True)
for i, zs in enumerate(zss[:4]):
    L, sbl = np.loadtxt(D+'sim_nres12_zs'+str(zs)+'_l200-2000_t0.0.dat',unpack=True,usecols=(0,1))
    l, tbl = np.loadtxt(D+'Sl_RT_zs'+str(zs)+'_zn100_l200-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
    plot(L,L*sbl,color=c[i],alpha=0.3)
    plot(l,l*tbl,label='$z_s='+str(zs)+'$',lw=1,color=c[i])
legend(loc=0,ncol=2)
```

CMB lensing kappa map

```python
pl.plot_1dstyle(ylab=r'$L S_L$',xmin=100,xmax=2000,ymin=8e-10,ymax=2e-7,ylog=True)
L, sbl = np.loadtxt(D+'sim_nres13_zs1100_l100-2000_t0.0.dat',unpack=True,usecols=(0,1))
l, tbl = np.loadtxt(D+'Sl_RT_zs1100_zn50_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
b, pbl = np.loadtxt(D+'Sl_RT_zs1100_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1))
plot(L,L*sbl,color='r',alpha=0.3)
plot(l,l*tbl,label='zn=50',lw=2,color='r')
plot(b,b*pbl,label='zn=100',lw=2,color='m')
legend(loc=0)
```

Gaussian smoothing

```python
for zi in [0,1,2,3,4]:
    pl.plot_1dstyle(ylab=r'$ratio$',fsize=[7,2.5],xmin=100,xmax=2000,ymin=0.5,ymax=1.5)
    for i, theta in enumerate([10.,5.,2.,0.]):
        L, sbl = np.loadtxt(D+'sim_nres13_zs'+str(zss[zi])+'_l100-2000_t'+str(theta)+'.dat',unpack=True,usecols=(0,1))
        l, tbl = np.loadtxt(D+'Sl_RT_zs'+str(zss[zi])+'_zn100_l100-2000_t'+str(theta)+'_pb.dat',unpack=True,usecols=(0,1))
        plot(l,sbl[np.array(l,dtype=np.int)]/tbl,label=r'$\theta='+str(theta)+'$',lw=2,color=c[i])
    legend(loc=0,ncol=2)
    axhline(1,color='k',ls='--')
    text(1000,0.8,'$z_s='+str(zss[zi])+'$')
    show()
    clf()
```

Changing Maximum multipole of convergence map

```python
for zi, ymin, ymax in [(0,1e-10,5e-9),(1,5e-10,2e-8),(2,1e-9,2e-8),(3,1e-9,3e-8),(4,1e-9,1e-7)]:
    pl.plot_1dstyle(ylab=r'$L\times S_L$',fsize=[7,4],ylog=True,xmin=100,xmax=2000,ymin=ymin,ymax=ymax)
    for i, theta in enumerate([10.,5.,2.,0.]):
        L, sbl = np.loadtxt(D+'sim_nres13_zs'+str(zss[zi])+'_l100-3000_t'+str(theta)+'.dat',unpack=True,usecols=(0,1))
        l, tbl = np.loadtxt(D+'Sl_RT_zs'+str(zss[zi])+'_zn100_l100-3000_t'+str(theta)+'_pb.dat',unpack=True,usecols=(0,1))
        b, Tbl = np.loadtxt(D+'Sl_RT_zs'+str(zss[zi])+'_zn100_l100-2000_t'+str(theta)+'_pb.dat',unpack=True,usecols=(0,1))
        plot(L,L*sbl,label='',color=c[i],alpha=0.3)
        plot(l,l*tbl,label=r'$\theta='+str(theta)+'$',lw=2,color=c[i])
        plot(b,b*Tbl,label='',lw=1,ls='--',color=c[i])
    legend(loc=0,ncol=2)
    figtext(0.2,0.9,'Dashed: $L_{max}=2000$')
    show()
    clf()
```
