Plot cross-skew spectra

```python
# external modules
import numpy as np
from matplotlib.pyplot import *
# moduels from cmblensplus/utils/
import plottools as pl
```

Initialize parameters

```python
D = 'data/skewspec/'
#zss = [0.5078,1.0334,1.5345,2.0548,1100]
zss = [0.5078]
c = ['r','g','b','m','c']
```

Plot both theoretical and simulated skew-spectra

```python
for i, zs0 in enumerate(zss):
    pl.plot_1dstyle(ylab=r'$L\times S_L$',xmin=100,xmax=2000,ymin=8e-10,ymax=7e-8,ylog=True)
    for j, zs1 in enumerate(zss):
        b, s0, s1, s2 = np.loadtxt(D+'Sl_RT_zs'+str(zs0)+'_zn100_l100-2000_t0.0.dat',unpack=True,usecols=(0,1,2,3))
        b, S0, S1, S2 = np.loadtxt(D+'Sl_RT_zs'+str(zs0)+'_zn100_l100-2000_t0.0_pb.dat',unpack=True,usecols=(0,1,2,3))
        plot(b,b*s0,color=c[j])
        plot(b,b*s1/b**2/3,color=c[j+1])
        plot(b,b*s2/b**4/3,color=c[j+1])
    show()
    clf()
```
