```python
import basic, numpy as np, healpy as hp, os, plottools as pl
from matplotlib.pyplot import *
from scipy.interpolate import InterpolatedUnivariateSpline as spline
```

```python
cpmodel = 'modelw'
model = 'RT'
D = '../data/2023_Farren+/'
k, pk0 = np.loadtxt(D+'../'+cpmodel+'/Pk/Pklin.dat',unpack=True)
kn = np.size(k)
zcmb = 1088.69
zm   = 1.
zs   = [zcmb,zcmb]
zmin, zmax = 0.01, 5.
zn = 30
btype = 'kgg'
lmin, lmax = 20, 3000
olmin, olmax = lmin, 2048
bn = lmax-lmin+1
ols = np.linspace(lmin,lmax,lmax-lmin+1,dtype=np.int)
z, dz = basic.bispec.zpoints(zmin,zmax,zn)
```

```python
zi, dndzi = np.loadtxt(D+'dNdz/blue.txt',unpack=True)
```

```python
dNdz = spline(zi,dndzi)(z)
imax = np.argmax(z>np.max(zi))
dNdz[imax:] = 0
```

```python
xlim(np.min(zi),np.max(zi))
plot(zi,dndzi)
plot(z,dNdz)
```
