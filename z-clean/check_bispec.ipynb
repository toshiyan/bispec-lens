```python
import numpy as np, local, basic
from matplotlib.pyplot import *
cycle = rcParams['axes.prop_cycle'].by_key()['color']
```

```python
kk_noise = 's4'
droot = 'data/modelw/snr/'
lmin, lmax = 1, 2048
L = np.linspace(0,lmax,lmax+1)
Lb = L[1:]
cpmodel = 'modelw'
model = 'RT'
```

```python
zn = 50
zmin, zmax = 0.0001, 20
z, dz = basic.bispec.zpoints(zmin,zmax,zn)
chi = basic.cosmofuncs.dist_comoving(z,**local.cps)
```

```python
k, pk0 = np.loadtxt(local.root + 'modelw/Pk/Pklin_new.dat',unpack=True)
# source redshift
zs = [local.zcmb,local.zcmb,local.zcmb]
# mass spectra
ckk, ckI, cII, nlII = np.loadtxt(local.root+'modelw/cl/mass.dat',unpack=True,usecols=(1,2,3,4))
fdel = ckI/(cII+nlII+1e-30)
# CIB weight
wcib = local.cib_weight(z,local.cps,local.nu)
# modified kernel
wdel = np.zeros((zn,lmax+1))
for l in range(lmax+1):
    if l<100: continue #low-ell cut
    wdel[:,l] = fdel[l]*wcib
# clkk
skk = basic.bispec.cl_flat(local.cpmodel,z,dz,zs[:2],lmax,k,pk0)
dkk = basic.bispec.cl_flat(local.cpmodel,z,dz,zs[:2],lmax,k,pk0,wdel=wdel)
```

```python
nlkk = np.zeros((lmax+1))
nlkk[2:] = np.loadtxt(local.root+'nldd/kappa_deproj0_sens0_16000_lT30-3000_lP30-5000.dat',unpack=True)[7][:lmax-1]
ockk = skk[:lmax+1] + nlkk
dckk = dkk[:lmax+1] + nlkk
```

```python
bl, BL, pb, PB = {}, {}, {}, {}
shapes = ['equi','fold','sque','angl']
```

```python
for shape in shapes:
    bl[shape], pb[shape] = basic.bispec.bispeclens(shape,cpmodel,model,z,dz,zs,lmin,lmax,k,pk0)
    BL[shape], PB[shape] = basic.bispec.bispeclens(shape,cpmodel,model,z,dz,zs,lmin,lmax,k,pk0,wdel=wdel)
```

```python
yscale('log')
ylim(1e-11,1e-8)
for c, shape in enumerate(shapes):
    nonzero = np.nonzero(bl[shape])
    Lb_plot = Lb[nonzero]
    bl_plot = bl[shape][nonzero]
    BL_plot = BL[shape][nonzero]
    plot(Lb_plot,Lb_plot**2*bl_plot,color=cycle[c],label=shape)
    plot(Lb_plot,Lb_plot**2*BL_plot,color=cycle[c],ls='--')
legend()
```

```python
l0 = int(lmin+(lmax-lmin)*0.5/20.)
ylim(1e-5,1e-2)
yscale('log')
print(l0)
eps = 1e-30
for c, shape in enumerate(shapes):
    if shape=='equi':
        snr = bl[shape]/(ockk[1:]+eps)**1.5
        SNR = bl[shape]/(dckk[1:]+eps)**1.5
    if shape=='fold':
        snr = np.array( [ bl[shape][l-1]/np.sqrt(ockk[l]*ockk[int(l/2)]**2+eps**3) for l in range(1,lmax+1) ] )
        SNR = np.array( [ BL[shape][l-1]/np.sqrt(dckk[l]*dckk[int(l/2)]**2+eps**3) for l in range(1,lmax+1) ] )
    if shape=='sque':
        snr = np.array( [ bl[shape][l-1]/np.sqrt(ockk[l0]*ockk[l]**2+eps**3) for l in range(1,lmax+1) ] )
        SNR = np.array( [ BL[shape][l-1]/np.sqrt(dckk[l0]*dckk[l]**2+eps**3) for l in range(1,lmax+1) ] )
    if shape=='angl':
        snr = np.array( [ bl[shape][l-1]/np.sqrt(ockk[l]*ockk[int(lmax/2)]**2+eps**3) for l in range(1,lmax+1) ] )
        SNR = np.array( [ BL[shape][l-1]/np.sqrt(dckk[l]*dckk[int(lmax/2)]**2+eps**3) for l in range(1,lmax+1) ] )
    nonzero = np.nonzero(bl[shape])
    Lb_plot = Lb[nonzero]
    snr = snr[nonzero]
    SNR = SNR[nonzero]
    plot(Lb_plot,snr,color=cycle[c],label=shape)
    plot(Lb_plot,SNR,color=cycle[c],ls='--')
legend()
```
