```python
import numpy as np, local
from matplotlib.pyplot import *
cycle = rcParams['axes.prop_cycle'].by_key()['color']
```

```python
kk_noise = 's4'
droot = 'data/modelw/snr/'
zn = 50
fsky = .4
```

```python
xlabel(r'$L_{\rm max}$')
ylabel('SNR')
ylim(0,20)
for i, zb in enumerate([6,7]):
    zbin = local.galaxy_distribution( np.linspace(0,50,1000), zbn={'lss':zb})[0]
    Lmaxs, oSNR, dSNR = np.loadtxt(
        local.root+local.cpmodel+'/snr/snr_kkk_zclean_lss_'+kk_noise+'_zbin'+str(zb)+'_zn'+str(zn)+'.dat',unpack=True,usecols=(0,3,4)
    )
    if i==0: plot(Lmaxs,oSNR*np.sqrt(fsky),ls='--',color='k',label='no cleaning')
    plot(Lmaxs,dSNR*np.sqrt(fsky),ls='-',color=cycle[i],label='cleaning, zmin='+str(zbin['lss'][0]))
legend()
savefig('fig_s4_snr_lss.png')
```
