```python
import numpy as np, local, basic
from matplotlib.pyplot import *
cycle = rcParams['axes.prop_cycle'].by_key()['color']
```

```python
lmin, lmax = 1, 2048
L = np.linspace(0,lmax,lmax+1)
Lb = L[1:]
cpmodel = 'modelw'
model = 'RT'
```

```python
zmaxs = np.arange(.3,10.,.3)
zmaxs = np.array([.3,.5,1.,2.,3.,5.])
```

```python
zmaxs
```

```python
k, pk0 = np.loadtxt(local.root + 'modelw/Pk/Pklin_new.dat',unpack=True)
# source redshift
zs = [local.zcmb,local.zcmb,local.zcmb]
zmin = 0.0001
```

```python
bl, pb = {}, {}
shapes = ['equi','fold','sque','angl']
```

```python
for shape in shapes:
    bl[shape] = np.zeros((len(zmaxs),lmax))
    pb[shape] = np.zeros((len(zmaxs),lmax))
    for zi, zmax in enumerate(zmaxs):
        zn = int(10*(zmax-zmin))
        z, dz = basic.bispec.zpoints(zmin,zmax,zn)
        chi = basic.cosmofuncs.dist_comoving(z,**local.cps)
        bl[shape][zi,:], pb[shape][zi,:] = basic.bispec.bispeclens(shape,cpmodel,model,z,dz,zs,lmin,lmax,k,pk0)
```

```python
yscale('log')
ylim(1e-11,1e-8)
for c, shape in enumerate(shapes):
    for zi, zmax in enumerate(zmaxs):
        BL = bl[shape][zi,:]
        nonzero = np.nonzero(BL)
        Lb_plot = Lb[nonzero]
        bl_plot = BL[nonzero]
        plot(Lb_plot,Lb_plot**2*bl_plot,color=cycle[zi],label=str(zmax))
    legend()
    show()
    clf()
```
